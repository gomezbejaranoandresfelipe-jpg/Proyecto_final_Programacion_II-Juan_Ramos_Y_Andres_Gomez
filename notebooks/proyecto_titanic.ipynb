```python
# 1. Crear la estructura obligatoria de la rúbrica
!mkdir -p data notebooks src outputs/tablas outputs/graficos docs

# 2. Descargar los datos oficiales por código
import requests

def descargar_csv(url, ruta_destino):
    respuesta = requests.get(url)
    with open(ruta_destino, 'wb') as archivo:
        archivo.write(respuesta.content)
    print(f"Descargado con éxito en: {ruta_destino}")

# Enlace al set de datos oficial
descargar_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv", "data/train.csv")
```

```python
import pandas as pd
import numpy as n  # Importación con tu alias estándar

# 1. Cargar el archivo descargado
df_train = pd.read_csv('data/train.csv')

# 4. Crear la variable 'Familiares' (Punto 4)
df_train['Familiares'] = df_train['SibSp'] + df_train['Parch']

# 5. Definir el dataframe unificado para las consultas del taller
df_full = df_train.copy()

print("¡DATOS PROCESADOS CON ÉXITO!\n")

# ---- MUESTRA DE RESULTADOS PARA TU INFORME ----
print(f"¿Cuál es la edad promedio?: {df_full['Age'].mean():.2f} años")
print(f"¿Cuál es la tarifa promedio en 1ra clase?: {df_full[df_full['Pclass'] == 1]['Fare'].mean():.2f}")

print("\n--- TASA DE SUPERVIVENCIA POR GÉNERO (%) ---")
print(df_full.groupby('Sex')['Survived'].mean() * 100)

print("\n--- TASA DE SUPERVIVENCIA POR CLASE (%) ---")
print(df_full.groupby('Pclass')['Survived'].mean() * 100)
```
