### CODIGO PARA EMPEZAR EL TALLER

```python
import pandas as pd
import numpy as n  # Importación estándar

# 1. Carga de datos independientes
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')

# 4. Creación de la variable 'Familiares' (Punto 4 del taller)
df_train['Familiares'] = df_train['SibSp'] + df_train['Parch']
df_test['Familiares'] = df_test['SibSp'] + df_test['Parch']

# 5. Preparación para la unión vertical (Punto 5 del taller)
df_train['source'] = 'train'
df_test['source'] = 'test'

# Concatenación vertical manteniendo los NaN en 'Survived' para el set de test
df_full = pd.concat([df_train, df_test], ignore_index=True)

print("¡Estructura unificada con éxito!")
print(f"Tamaño total del DataFrame unificado: {df_full.shape}")
```

### PUNTO 2 Y 3

```python
# Ver tipos de datos y nulos globales
print("--- INFORMACIÓN DEL DATASET ---")
df_full.info()

# Estadísticas de variables numéricas en train y test para comparar
print("\n--- ESTADÍSTICAS TABULAR TRAIN ---")
print(df_train.describe())

print("\n--- ESTADÍSTICAS TABULAR TEST ---")
print(df_test.describe())
```

### PUNTO 7

```python
# Conteo de datos faltantes por columna
valores_nulos = df_full.isnull().sum()
print("Datos faltantes en el conjunto unificado:")
print(valores_nulos[valores_nulos > 0])
```

### PUNTO 8

```python
print(f"(a) Edad promedio de los pasajeros: {df_full['Age'].mean():.2f} años")
sobrevivientes = df_train['Survived'].value_counts()
print(f"(b) Pasajeros que murieron (0): {sobrevivientes[0]} | Sobrevivieron (1): {sobrevivientes[1]}")
tarifa_primera = df_full[df_full['Pclass'] == 1]['Fare'].mean()
print(f"(c) Tarifa promedio en primera clase: {tarifa_primera:.2f}")
con_familiar = len(df_full[df_full['Familiares'] > 0])
print(f"(d) Pasajeros que viajaron con al menos un familiar: {con_familiar}")
print(f"(e) Edad más joven: {df_full['Age'].min()} | Edad más vieja: {df_full['Age'].max()}")
print("\n(f) Pasajeros por puerto de embarque:")
print(df_full['Embarked'].value_counts())
solos = len(df_full[df_full['Familiares'] == 0])
con_familia = len(df_full[df_full['Familiares'] > 0])
print(f"\n(g) Pasajeros que viajaron solos: {solos} | Con familiares: {con_familia}")
```

### PUNTO 9

```python
print("--- (a) Supervivencia por Sexo ---")
print(df_train.groupby('Sex')['Survived'].agg(['count', 'mean']))
print("\n--- (b) Supervivencia Niños (<18) vs Hombres Adultos ---")
df_train['es_nino'] = df_train['Age'] < 18
df_train['es_hombre_adulto'] = (df_train['Sex'] == 'male') & (df_train['Age'] >= 18)
print("Tasa de supervivencia Niños (<18):", df_train[df_train['es_nino']]['Survived'].mean())
print("Tasa de supervivencia Hombres Adultos:", df_train[df_train['es_hombre_adulto']]['Survived'].mean())
print("\n--- (d) Supervivencia por Puerto de Embarque ---")
print(df_train.groupby('Embarked')['Survived'].agg(['count', 'mean']))
print("\n--- (e) Supervivencia por Clase de Tiquete ---")
print(df_train.groupby('Pclass')['Survived'].agg(['count', 'mean']))
```

### PUNTO 10

```python
# Clasificar los grupos familiares
def clasificar_familia(cant):
    if cant == 0:
        return 'Grupo 1: Solo'
    elif 1 <= cant <= 3:
        return 'Grupo 2: Familia Pequeña'
    else:
        return 'Grupo 3: Familia Grande'

df_full['Grupo_Familiar'] = df_full['Familiares'].apply(clasificar_familia)
# Crear indicador de si tiene cabina asignada (no nula)
df_full['Tiene_Cabina'] = df_full['Cabin'].notnull()
# Mostrar la distribución porcentual cruzada
dist_cabinas = pd.crosstab(df_full['Grupo_Familiar'], df_full['Tiene_Cabina'], normalize='index') * 100
print(dist_cabinas)
```

### PUNTO 11

```python
# 1. Definir los grupos etarios obligatorios
def segmentar_edad(edad):
    if edad < 10:
        return 'Niño'
    elif 18 <= edad <= 49:
        return 'Adulto'
    elif edad >= 50:
        return 'Mayor'
    else:
        return 'Adolescente / No Registrado'
df_train['grupo_etario'] = df_train['Age'].apply(segmentar_edad)
df_train['Tiene_Cabina'] = df_train['Cabin'].notnull()

# 2. Construir la tabla multi-índice con las tasas de supervivencia relativas
tabla_analitica = df_train.groupby(['Sex', 'grupo_etario', 'Tiene_Cabina'])['Survived'].mean().unstack() * 100
print("Tasas de supervivencia (%) por categoría cruzada:")
print(tabla_analitica)
```
