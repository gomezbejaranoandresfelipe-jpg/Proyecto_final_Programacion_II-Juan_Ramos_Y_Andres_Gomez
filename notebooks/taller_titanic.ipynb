### PUNTOS 1, 2 Y 3

```python
import pandas as pd
import numpy as n

# 1. Carga de los datos oficiales
df_train = pd.read_csv('data/train.csv')

# 2. Definición del DataFrame unificado base para el taller
df_full = df_train.copy()

# 3. Inspección inicial de tipos de datos y dimensiones
print("--- PUNTO 3: ESTRUCTURA DEL DATASET ---")
print(f"Dimensiones del DataFrame: {df_full.shape}")
print("\nTipos de datos por columna:")
print(df_full.dtypes)
```

### PUNTOS 4 Y 5

```python
# 4. Creación de la variable 'Familiares' (Suma de hermanos/esposos y padres/hijos)
df_full['Familiares'] = df_full['SibSp'] + df_full['Parch']

# 5. Confirmación de la estructura unificada con la nueva columna
print("--- PUNTO 5: VERIFICACIÓN DEL DATAFRAME ---")
print(f"Tamaño actual del DataFrame: {df_full.shape}")
print(f"Columnas disponibles: {list(df_full.columns)}")
```

### PUNTOS 6 Y 7

```python
print("--- PUNTO 6: CONTEO DE DATOS FALTANTES ---")
valores_nulos = df_full.isnull().sum()
print(valores_nulos[valores_nulos > 0])

# 7. Imputación estratégica de nulos
# Rellenar Edad con la mediana para no sesgar los grupos etarios
df_full['Age'] = df_full['Age'].fillna(df_full['Age'].median())

# Crear indicador booleano para la presencia de Cabina (True si tiene, False si es NaN)
df_full['Tiene_Cabina'] = df_full['Cabin'].notnull()

print("\nValores nulos después de la limpieza en Edad y Cabina:")
print(df_full[['Age', 'Tiene_Cabina']].isnull().sum())
```

### PUNTO 8

```python
print("--- PUNTO 8: MÉTRICAS ESTADÍSTICAS ---")
print(f"a) Edad promedio de los pasajeros: {df_full['Age'].mean():.2f} años")
print(f"b) Pasajeros que sobrevivieron: {df_full['Survived'].sum()}")
print(f"c) Tarifa promedio en Primera Clase: {df_full[df_full['Pclass'] == 1]['Fare'].mean():.2f}")
print(f"d) Edad del pasajero más joven: {df_full['Age'].min()} años | Más viejo: {df_full['Age'].max()} años")
print(f"e) Cantidad de pasajeros que viajaban solos: {len(df_full[df_full['Familiares'] == 0])}")
```

### PUNTO 9

```python
print("--- PUNTO 9: TASAS DE SUPERVIVENCIA ---")
print("\na) Por Sexo (%):")
print((df_full.groupby('Sex')['Survived'].mean() * 100).round(2))
print("\nb) Por Clase de Pasajero (%):")
print((df_full.groupby('Pclass')['Survived'].mean() * 100).round(2))
print("\nc) Por Puerto de Embarque (%):")
print((df_full.groupby('Embarked')['Survived'].mean() * 100).round(2))
```

### PUNTOS 10 Y 11

```python
# 10. Función para clasificar grupos familiares
def clasificar_familia(cant):
    if cant == 0: return 'Solo'
    elif cant <= 3: return 'Familia Pequeña'
    else: return 'Familia Grande'

df_full['Grupo_Familiar'] = df_full['Familiares'].apply(clasificar_familia)

# 11. Función para segmentar grupos etarios
def segmentar_edad(edad):
    if edad < 12: return 'Niño'
    elif edad <= 18: return 'Adolescente'
    elif edad <= 60: return 'Adulto'
    else: return 'Mayor'

df_full['Grupo_Etario'] = df_full['Age'].apply(segmentar_edad)

print("--- PUNTO 11: TABLA ANALÍTICA CRUZADA MULTI-ÍNDICE (%) ---")
tabla_analitica = df_full.groupby(['Grupo_Familiar', 'Grupo_Etario', 'Tiene_Cabina'])['Survived'].mean() * 100
print(tabla_analitica.unstack(level=-1).round(2))
```
